```python
import csv

# Open the CSV file
with open('clinical_data.csv', 'r') as csvfile:
    reader = csv.DictReader(csvfile)
    
    # Iterate over each row in the CSV file
    for row in reader:
        # Convert selected fields to appropriate data types
        row['methylation_clust'] = int(row['methylation_clust'])
        row['gebe_exp_clust'] = int(row['gebe_exp_clust'])
        row['Age_InitDiag'] = int(row['Age_InitDiag'])
        row['survive_code'] = int(row['survive_code'])
        row['OS_Time'] = int(row['OS_Time'])

        # Insert each modified row as a document into the MongoDB collection
        document = {
            'RefSeq_accession_number': row['RefSeq_accession_number'],
            'ID': row['ID'],
            'Gender': row['Gender'],
            'ER': row['ER'],
            'PR': row['PR'],
            'Node': row['Node'],
            'Node_Coded': row['Node_Coded'],
            'Metastasis': row['Metastasis'],
            'Metastasis_Coded': row['Metastasis_Coded'],
            'survive': row['survive'],            
            'PAM50_mRNA': row['PAM50_mRNA'],
            'methylation_clust': row['methylation_clust'],
            'gebe_exp_clust': row['gebe_exp_clust'],
            'Age_InitDiag': row['Age_InitDiag'],
            'HER2_FinalStatus': row['HER2_FinalStatus'],
            'survive_code': row['survive_code'],
            'OS_Time': row['OS_Time'],
            

    
        }
        collection2.insert_one(document)
        
print("Data has been successfully imported into MongoDB.")
```

```python

def insert_many_drugs():
    # Get the number of drugs to insert
    num_of_inserted_drugs = int(input("Enter number of drugs: "))
    
    # Prompt the user for data for each drug
    ids = []
    drug_names = []
    conditions = []
    reviews = []
    ratings = []
    dates = []
    useful_counts = []
    
    for i in range(num_of_inserted_drugs):
        uniqueID = input(f"Enter the ID of the new drug ({i+1}): ")
        drug_name = input(f"Enter the name of the new drug ({i+1}): ")
        condition = input(f"Enter the condition of the new drug ({i+1}): ")
        review = input(f"Enter the review of the new drug ({i+1}): ")
        rating = input(f"Enter the rating of the new drug ({i+1}): ")
        date = input(f"Enter the date of the new drug ({i+1}): ")
        useful_count = input(f"Enter the useful count of the new drug ({i+1}): ")
        
        ids.append(uniqueID)
        drug_names.append(drug_name)
        conditions.append(condition)
        reviews.append(review)
        ratings.append(rating)
        dates.append(date)
        useful_counts.append(useful_count)
    
    # Create a list of dictionaries representing the new documents
    drug_data = []
    for i in range(num_of_inserted_drugs):
        drug_data.append({
            'id': ids[i],
            'drugName': drug_names[i],
            'condition': conditions[i],
            'review': reviews[i],
            'rating': ratings[i],
            'date': dates[i],
            'usefulCount': useful_counts[i]
        })
    
    # Insert the new documents into the collection
    result = collection.insert_many(drug_data)
    print("New drugs added with IDs:", result.inserted_ids)

# Call the function to insert new drug documents
insert_many_drugs()
```
