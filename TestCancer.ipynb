```python
def insert_many_drugs():
    # Get the number of drugs to insert
    num_of_inserted_drugs = int(input("Enter number of drugs: "))
    
    # Prompt the user for data for each drug
    ids = []
    drug_names = []
    conditions = []
    reviews = []
    ratings = []
    dates = []
    useful_counts = []
    
    for i in range(num_of_inserted_drugs):
        uniqueID = input(f"Enter the ID of the new drug ({i+1}): ")
        drug_name = input(f"Enter the name of the new drug ({i+1}): ")
        condition = input(f"Enter the condition of the new drug ({i+1}): ")
        review = input(f"Enter the review of the new drug ({i+1}): ")
        rating = input(f"Enter the rating of the new drug ({i+1}): ")
        date = input(f"Enter the date of the new drug ({i+1}): ")
        useful_count = input(f"Enter the useful count of the new drug ({i+1}): ")
        
        ids.append(uniqueID)
        drug_names.append(drug_name)
        conditions.append(condition)
        reviews.append(review)
        ratings.append(rating)
        dates.append(date)
        useful_counts.append(useful_count)
    
    # Create a list of dictionaries representing the new documents
    drug_data = []
    for i in range(num_of_inserted_drugs):
        drug_data.append({
            'id': ids[i],
            'drugName': drug_names[i],
            'condition': conditions[i],
            'review': reviews[i],
            'rating': ratings[i],
            'date': dates[i],
            'usefulCount': useful_counts[i]
        })
    
    # Insert the new documents into the collection
    result = collection.insert_many(drug_data)
    print("New drugs added with IDs:", result.inserted_ids)

# Call the function to insert new drug documents
insert_many_drugs()
```

```python
from pymongo import MongoClient

def update_field_as_int(db_name, collection_name, field_name):
    # Connect to the MongoDB server
    client = MongoClient()

    # Access the specified database and collection
    db = client[db_name]
    collection = db[collection_name]

    # Update the field values to integers using $toInt operator
    collection.update_many({}, [{"$set": {field_name: {"$toInt": f"${field_name}"}}}])

    # Close the MongoDB connection
    client.close()

# Example usage
db_name = 'your_database_name'
collection_name = 'your_collection_name'
field_name = 's7'

update_field_as_int(db_name, collection_name, field_name)
```
